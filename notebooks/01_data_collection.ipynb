```python
import pandas as pd

#Load crime dataset
crime = pd.read_csv('../data/raw/crime.csv')

#First Look at Data
print(crime.shape)
print(crime.columns.tolist())
```

```python
crime.isnull().sum()
```

```python
crime = crime.drop(columns=['INCIDENT_NUMBER', 'OFFENSE_CODE', 'REPORTING_AREA', 'STREET', 'DAY_OF_WEEK', 'HOUR', 'Location', 'YEAR', 'MONTH'])
```

```python
print(crime.shape)
crime.head()
```

```python
crime = crime[crime['YEAR'] >= 2021]
print(crime.shape)
```

```python
crime['YEAR'].value_counts().sort_index()
```

```python
crime = crime.drop(columns=['YEAR', 'MONTH'])
print(crime.shape)
crime.head()
```
